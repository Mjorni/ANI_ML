```python
import pandas
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pandas.read_csv("AnimeData.csv")
```

```python
# Целый датафрейм 1
data
```

```python
#Средний рейтинг
data["Реитинг"].mean()
```

```python
# Получение данных о типе и их количестве
type_count = data["Тип"].value_counts()
type_count
type_unique = type_count.index.tolist()
type_count
```

```python
# Отрисовка графика количества различных типов
#Первый бар
plt.bar(type_unique, type_count, width=0.7, edgecolor="white", linewidth=0.7)
plt.show()
```

```python
# сортировка от мусора и разбиение по возростному рейтингу
data = data[data["Возростной рейтинг"]!= " /"]

pg_count = data["Возростной рейтинг"].value_counts()
pg_unique = pg_count.index.tolist()
pg_count
```

```python
#бар по воз.рейтингу
#Второй бар
plt.bar(pg_unique, pg_count, width=1, edgecolor="white", linewidth=0.7)
plt.show()
```

```python
# Приводим данные к значению 1 или 0 и выводим в виде бара 
status_count = data["Статус"].replace([1,0], ["Вышел", "Не вышел"]).value_counts()
status_unique = status_count.index
status_unique
```

```python
# Вывод бара
#Третий бар
plt.bar(status_unique, status_count, width=1, edgecolor="white", linewidth=0.7)
plt.show()
```

```python
# вывод бара по колонце "оценивают"
pg_count = data["Оценивают"].value_counts()
pg_unique = pg_count.index.tolist()
pg_count
```

```python
#Четвертый бар
plt.bar(pg_unique, pg_count, width=1, edgecolor="white", linewidth=0.7)
plt.show()
```

```python
# Функция которая отвечает за топ по заданым параметрам
#    column - название стобца
#    т - количество столбцов
#    bar - если false то отрисовывается датафрейм, в ином случае рисуется бар

def get_n_most_big(column, n,  bar = False):
    if bar:
        var_df = data.sort_values(by=column , ascending=False)[0:n]
        name = var_df['Название'].tolist()
        second_var = var_df[column]
        ax = sns.barplot(x = second_var, y = name)
        for i in ax.containers:
            ax.bar_label(i, color = "white", label_type = "center")
    else:
        return data.sort_values(by=column , ascending=False)[0:n]
```

```python
get_n_most_big("Реитинг", 20, bar = False)
```

```python
#Датафрейм 3
get_n_most_big("Реитинг", 10, bar = False)
```

```python
#Датафрейм 4
get_n_most_big("Рецензий", 10, bar = False)
```

```python
#Датафрейм 5
get_n_most_big("Количество", 10, bar = False)
```

```python
#Датафрейм 5
get_n_most_big("Количество", 10, bar = True)
```
